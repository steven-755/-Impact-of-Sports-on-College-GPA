```python
df = pd.read_csv("Untitled form (Responses) - Form Responses 1.csv")
df.head(30)
```

```python
maping = {
    "1-2 hours per week": 1.5,
    "2-4 hours per week": 3,
    "4-6 hours per week": 5,
    "More than 6 hours per week": 7
}
df['studying time'] = df['On average, how many hours per week do you spend studying ?'].map(maping)
```

```python
mapping = {
    "No, or not on a regular basis": 0,
    "Yes, 1-2 hours per week": 1.5,
    "Yes, 2-4 hours per week": 3,
    "Yes, 4-6 hours per week": 5
}
df['Sports Participation'] = df['Do you participate in any sports or physical activities regularly ?'].map(mapping)
```

```python
df.head(30)
```

```python

# Grouped bar chart of GPA by belief about the positive impact of sports participation
plt.figure(figsize=(8, 6))
df.groupby('Do you believe sports participation positively impacts academic performance ?')['What is your current GPA?'].mean().plot(kind='bar', color='skyblue')
plt.xlabel('Belief about the Positive Impact of Sports Participation')
plt.ylabel('Average GPA')
plt.title('Average GPA by Belief about Sports Participation Impact')
plt.xticks(rotation=45)
plt.show()
```

```python

# Create a pivot table to count the frequency of each belief category for each level of sports participation
belief_vs_sports = pd.pivot_table(df, index='Do you believe sports participation positively impacts academic performance ?', columns='Do you participate in any sports or physical activities regularly ?', aggfunc='size', fill_value=0)

# Plot the stacked bar chart
belief_vs_sports.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightblue', 'lightgreen', 'salmon', 'skyblue'])
plt.xlabel('Belief about the Positive Impact of Sports Participation')
plt.ylabel('Frequency')
plt.title('Belief vs. Sports Participation')
plt.xticks(rotation=45)
plt.legend(title='Sports Participation', loc='upper right')
plt.show()
```

```python
# Scatterplot of GPA vs. Sports Participation
plt.figure(figsize=(8, 6))
plt.scatter(df['Sports Participation'], df['What is your current GPA?'], color='purple')
plt.xlabel('Sports Participation time')
plt.ylabel('GPA')
plt.title('GPA vs. Sports Participation')
plt.grid(True)
plt.show()
```

```python
# Scatterplot of GPA vs. Sports Participation
plt.figure(figsize=(8, 6))
plt.scatter(df['Sports Participation'], df['On average, how many hours per week do you spend studying ?'], color='purple')
plt.xlabel('Sports Participation time')
plt.ylabel('Studying time')
plt.title('Studying time vs. Sports Participation')
plt.grid(True)
plt.show()
```

```python
# Calculate mean, median, and mode for GPA
mean_gpa = df['What is your current GPA?'].mean()
median_gpa = df['What is your current GPA?'].median()
mode_gpa = df['What is your current GPA?'].mode()[0]  # If there are multiple modes, take the first one

# Calculate mean, median, and mode for sports participation
mean_sports = df['Sports Participation'].mean()
median_sports = df['Sports Participation'].median()
mode_sports = df['Sports Participation'].mode()[0]  # If there are multiple modes, take the first one

print("Mean GPA:", mean_gpa)
print("Median GPA:", median_gpa)
print("Mode GPA:", mode_gpa)

print("Mean Sports Participation:", mean_sports)
print("Median Sports Participation:", median_sports)
print("Mode Sports Participation:", mode_sports)
```
